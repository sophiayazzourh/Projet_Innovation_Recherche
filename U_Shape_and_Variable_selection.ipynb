# Libraries and data

```python
## Libraries 
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import matplotlib.pyplot as plt

### Warning 
import warnings
warnings.filterwarnings("ignore")
```

```python
## Data
data = pd.read_excel("./Base_PBM_Facteurs_Confusion.xlsx")
### Data cleaning 
data["transfusion CG"][np.where(data["transfusion CG"] > 0)[0]] = 1
data.dropna(inplace=True)
data.set_index(np.arange(len(data)))
print(len(data))
data.head()
```

```python
print(data.columns)
len(data.columns)
```

# Facteurs de confusion 

```python
## Exactraction 
Facteurs_Confusion = data.iloc[:,:-3]
X = Facteurs_Confusion 
y = data["transfusion CG"]
## Model 
from sklearn.linear_model import LogisticRegression
logitS = LogisticRegression()
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
## Backward 
# Sequential Backward Selection
sbs = SFS(logitS, 
          forward=False, 
          floating=False, 
          scoring='r2', )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
```
