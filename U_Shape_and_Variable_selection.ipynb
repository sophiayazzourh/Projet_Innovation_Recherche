# Libraries and data

# WITHOUT SPLINES PRE PROCESSING 

```python
## Data
data = pd.read_excel("./Base_PBM_Facteurs_Confusion.xlsx")
### Data cleaning 
data["transfusion CG"][np.where(data["transfusion CG"] > 0)[0]] = 1
data.dropna(inplace=True)
data.set_index(np.arange(len(data)))
print(len(data))
data.head()
```

# Facteurs de confusion 

```python
Facteurs_Confusion = data.copy()
```

```python
## Supp no-needed columns
del Facteurs_Confusion['fibrinogene']
del Facteurs_Confusion['transfusion CG']
X = Facteurs_Confusion 
y = data["transfusion CG"]
## Model 
from sklearn.linear_model import LogisticRegression
logitS = LogisticRegression()
```

```python
Facteurs_Confusion.columns
```

### Librairie 1 

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
## Backward 
# Sequential Backward Selection AVEC R2
sbs = SFS(logitS, 
          forward=False, 
          floating=False, 
          scoring='r2' )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
print(sbs.k_feature_names_)
```

```python
#sbs.subsets_
```

```python
## Backward 
# Sequential Backward Selection AVEC NEG MEAN SQUARED ERROR
sbs = SFS(logitS, 
          forward=False,
          scoring='neg_mean_squared_error' )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
print(sbs.k_feature_names_)
```

```python
## Backward 
# Sequential Backward Selection AVEC ACCURACY
sbs = SFS(logitS, 
          forward=False,
          scoring='accuracy' )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
print(sbs.k_feature_names_)
```

```python
## Forward
# Sequential Forward Selection AVEC R2
sbs = SFS(logitS,
          scoring='r2' )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
print(sbs.k_feature_names_)
```

```python
### STEPWISE
sbs = SFS(logitS,
          floating=True,
          scoring='r2' )
sbs = sbs.fit(X, y)

print('\nSequential Backward Selection :')
print(sbs.k_feature_idx_)
print('CV Score:')
print(sbs.k_score_)
print(sbs.k_feature_names_)
```

### Librairie 2 : sklearn

https://scikit-learn.org/stable/modules/feature_selection.html
    
L1-based feature selection
Linear models penalized with the L1 norm have sparse solutions: many of their estimated coefficients are zero. When the goal is to reduce the dimensionality of the data to use with another classifier, they can be used along with feature_selection.SelectFromModel to select the non-zero coefficients. In particular, sparse estimators useful for this purpose are the linear_model.Lasso for regression, and of linear_model.LogisticRegression and svm.LinearSVC for classification.

```python
from sklearn.feature_selection import SelectFromModel
logitS = LogisticRegression().fit(X,y)
model = SelectFromModel(logitS, prefit=True)
X_new = model.transform(X)
```

```python
X_new.shape
```

```python
feature_names = np.array(X.columns)
print("Features selected by SelectFromModel: "
      f"{feature_names[model.get_support()]}")
```

### Solution 3 avec Statmodel 

```python
import statsmodels.api as sm
```

```python
#Adding constant column of ones, mandatory for sm.OLS model
X_1 = sm.add_constant(X)
#Fitting sm.OLS model
model = sm.OLS(y.astype(float),X_1.astype(float)).fit()
model.pvalues
```

```python
#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p=[]
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y.astype(float),X_1.astype(float)).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

# WITH SPLINE 

```python
from patsy import dmatrix 
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
## Cubic spline (B_Spline) process on fibrinogene rdb
transformed_fibri = dmatrix("bs(train, knots=(1.5, 2, 2.5, 3, 3.5), degree=3, include_intercept=False)", {"train": data['Fibrinogène rdb']},return_type='dataframe')
Facteurs_Confusion_Spline = pd.concat([data.copy(),transformed_fibri], axis = 1)
```

```python
transformed_fibri.head()
```

```python
## Supp no-needed columns
del Facteurs_Confusion_Spline['fibrinogene']
del Facteurs_Confusion_Spline['transfusion CG']
del Facteurs_Confusion_Spline['Fibrinogène rdb']
del Facteurs_Confusion_Spline['Intercept']
X = Facteurs_Confusion_Spline
y = data["transfusion CG"]
## Model 
from sklearn.linear_model import LogisticRegression
logitS = LogisticRegression()
```

```python
Facteurs_Confusion_Spline.columns
```

### sklearn

```python
from sklearn.feature_selection import SelectFromModel
logitS = LogisticRegression().fit(X,y)
model = SelectFromModel(logitS, prefit=True)
X_new = model.transform(X)
```

```python
feature_names = np.array(X.columns)
print("Features selected by SelectFromModel: "
      f"{feature_names[model.get_support()]}")
```
