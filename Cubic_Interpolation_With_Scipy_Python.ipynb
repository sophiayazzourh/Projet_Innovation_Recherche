## Spline interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

### 1st MÃ©thode : return interpolation fuctions for data in 1D 

```python
from scipy.interpolate import interp1d
```

interp1d : x and y are arrays of values used to approximate some function f: y = f(x). This class returns a function whose call method uses interpolation to find the value of new points.



```python
# Data
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/9.0)
# Interpolations
f = interp1d(x, y) # Linear spline function creation
f2 = interp1d(x, y, kind='cubic') # Cubic spline function creation
```

```python
# Number of point asking for the plot
# Our original space is between 0 and 10 and compose of 11 points 
# In order to plot our interpolation function, we decide to use 41 points between 0 and 1
xnew = np.linspace(0, 10, num=41, endpoint=True)

# Plot
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```

### 2nd MÃ©thode :  B -Spline 

B-Spline : https://fr.wikipedia.org/wiki/B-spline

splred : 
Find the B-spline representation of a 1-D curve.

We can parametrise it such as determine de spline on interval [xb, xe]. By default, xb = x[0] and xe = x[-1]. 

```python
x = np.arange(0, 2*np.pi+np.pi/4, 2*np.pi/8)
y = np.sin(x)
tck = interpolate.splrep(x, y, s = 0)
xnew = np.arange(0, 2*np.pi, np.pi/50)
ynew = interpolate.splev(xnew, tck, der=0)
```

```python
plt.figure()
plt.plot(x, y, 'X', xnew, ynew, '-')
plt.legend(['Data', 'Cubic Spline'])
plt.axis([-0.05, 6.33, -1.05, 1.05])
plt.title('Cubic-spline interpolation')
plt.show()
```
